```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import nltk
import json
```

ISEAR dataset

```python
# load data
from py_isear.isear_loader_spacy import IsearLoader
attributes = []
target = ['EMOT']
loader = IsearLoader(attributes, target, True)
data = loader.load_isear('isear.csv')
```

```python
# sample data
print(data.get_freetext_content()[1])
print(data.get_target()[1])
```

```python
# label occurrences
print(pd.value_counts(data.get_target()))
"""
1 JOY
2 FEAR
3 ANGER
4 SADNESS
5 DISGUST
6 SHAME
7 GUILT
"""
# token occurrences
all_tokens = [element for lis in data.get_freetext_content() for element in lis]
Freq_dist_nltk=nltk.FreqDist(all_tokens)
print(Freq_dist_nltk)
Freq_dist_nltk.plot(50,cumulative=False)
```

TEC dataset

```python
# load data
from py_tec.tec_loader_spacy import TecLoader
loader = TecLoader(True)
data = loader.load_tec('tec.txt')
```

```python
# sample data
print(data.get_data()[1])
print(data.get_target()[1])
```

```python
# label occurrences
print(pd.value_counts(data.get_target()))
"""
':: joy': 1,
':: fear': 2,
':: anger': 3,
':: sadness': 4,
':: disgust': 5,
':: surprise': 6
"""
# token occurrences
all_tokens = [element for lis in data.get_data() for element in lis]
Freq_dist_nltk=nltk.FreqDist(all_tokens)
print(Freq_dist_nltk)
Freq_dist_nltk.plot(50,cumulative=False)
labels = ['joy', 'fear', 'anger', 'sadness', 'disgust', 'surprise']
counts = json.load(open('py_tec/tec_feature_count.json'))
print([word[0] for word in counts])
fig1, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.pie(counts[4][1], labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=270)
ax1.axis('equal')
ax1.set_title(counts[4][0])
ax2.pie(counts[0][1], labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=270)
ax2.axis('equal')
ax2.set_title(counts[0][0])
plt.show()

```

SSEC dataset

```python
# load data
from py_ssec.ssec_loader import SsecLoader
loader = SsecLoader(True)
data = loader.load_ssec('ssec-aggregated/test-combined-0.0.csv')
```

```python
# sample data
print(data.get_data()[1])
print(data.get_target()[1])
```

```python
# label occurrences
print(np.sum(data.get_target(), axis = 0))
print(len(data.get_target()), ' in total')
# token occurrences
all_tokens = [element for lis in data.get_data() for element in lis]
Freq_dist_nltk=nltk.FreqDist(all_tokens)
print(Freq_dist_nltk)
Freq_dist_nltk.plot(50,cumulative=False)
```
